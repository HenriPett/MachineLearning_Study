# Transfer Learning

O transfer learning √© usar um modelo pr√©-treinado, por√©m para outro proposito o qual foi previamente definido. √â a reutiliza√ß√£o de um modelo, para um novo problema. Isto √©, vou usar uma rede neural treinada em outro outro conjunto de dados, geralmente maior, para resolver um novo problema.

O segredo √© treinar o seu modelo em um modelo pr√©-treinado, para conseguir uma acuracia melhor do que o comum.

Fine tunning √© o nome dessa t√©cnica, onde atualiza os pesos do modelo pr√©-treinado, adicionando novas "ep√≥cas" (epochs) com uma diferente tarefa.

Caso seja feito um fine tunning na resnet (que contem inumeras imagens aleat√≥rias de varios objetos), por exemplo, com mais imagens de gatos. O modelo ficar√° melhor em identificar gatos, por√©m pior em identificar outras coisas, pois ele acabou "esquecendo" um pouco das outras imagens. Uma boa solu√ß√£o para esse problema √©, continuar implementando imagens aleat√≥rias (ao invez de s√≥ gatos), porem em menor quantidade.

## Precis√£o

Fornece o numero de vezes que uma determinada classe foi predita corretamente dividido pelo numero de vezes que a classe foi predita.
